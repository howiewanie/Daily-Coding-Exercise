```python
#answer
def rotate(lst, k):
    for _ in range(k):
        # Move each element down by one.
        first_element = lst[0]
        for i in range(len(lst) - 1):
            lst[i] = lst[i + 1]
        lst[len(lst) - 1] = first_element
    return lst
```

```python
'''Write a function that rotates a list by k elements. For example, [1, 2, 3, 4, 5, 6] 
rotated by two becomes [3, 4, 5, 6, 1, 2].'''
def rotate(lst, n):
    outlst = []
    for i in range(0,len(lst)):
        if i + n < len(lst):
            outlst.append(lst[i+n])
        else:
            outlst.append(lst[i+n-len(lst)])
    return(outlst)
rotate([1,2,3,4,5,6],2)
```
