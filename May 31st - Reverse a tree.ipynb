```python
def rev(string):
    string=string[::-1]
    return string

def rev_tree(tree):
    tree = []
    row_index = 0
    for line in tree:
        if row_index % 2 == 1:
            newline = line
        else:
            newline = rev(line)
            tree.append(newline)
    return tree
```

```python
#answer

def reverse(self):
    self.left, self.right = self.right, self.left #this is declaring the variable
    if self.left != None: #new left is the old right
        self.left.reverse()
    if self.right != None:
        self.right.reverse()
    return self

#https://www.youtube.com/watch?v=-dTqUVkRTVg
```
