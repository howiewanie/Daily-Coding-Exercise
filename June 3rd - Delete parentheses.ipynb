```python
def count_del(s):
    invalid = 0
    open = 0
    for c in s:
        if c == '(':
            open += 1
        elif c == ')':
            if open > 0:
                open -= 1
            else:
                invalid += 1 
    #total = invalid + open
    return invalid + open
    
```

```python
def count_invalid_parenthesis(string):
    opened = 0
    invalid = 0
    for c in string:
        if c == '(':
            opened += 1
        elif c == ')':
            if opened > 0:
                opened -= 1
            else:
                invalid += 1
    # Count as invalid all unclosed parenthesis
    invalid += opened
    return invalid
```
