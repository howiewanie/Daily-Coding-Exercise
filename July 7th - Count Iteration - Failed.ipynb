```python
#Run-length encoding is a fast and simple method of encoding strings. The basic idea is to represent repeated successive characters 
#as a single count and character. 
#For example, the string "AAAABBBCCDAA" would be encoded as "4A3B2C1D2A".

#Implement run-length encoding and decoding. You can assume the string to be encoded have no digits and consists solely of 
#alphabetic characters. You can assume the string to be decoded is valid.
```

```python
def repeatcount(n):
    new_list = ''
    current_char = n[0]
    count = 1
    for i,char in enumerate(n,0):
        if char == current_char:
            count += 1          
        else:
            new_list += str(count) + current_char
            current_char = char
            count = 1
    new_list += str(count) + current_char
    return new_list
```

```python
repeatcount("AAAABBBCCDAA")
```

```python
def encode(s):
    result = ''
    current_char = s[0]
    current_count = 1
    for i, char in enumerate(s, 0):
        if char == current_char:
            current_count += 1
        else:
            result += str(current_count) + current_char
            current_char = char
            current_count = 1
    result += str(current_count) + current_char
    return result
```
