```python
#Given a function that generates perfectly random numbers between 1 and k (inclusive), 
#where k is an input, write a function that shuffles a deck of cards represented as an array using only swaps.
```

```python
from random import randint
def shuffle(card):
    k = len(card)
    for i in range(k):
        rand = randint(0, k-1)
        card[i], card[rand] = card[rand], card[i]
    return card
```

```python
shuffle(['a', 'b', 'c'])
```

```python
from random import randint

def shufflee(arr):
    n = len(arr)
    for i in range(n - 1):
        j = randint(i, n - 1)
        arr[i], arr[j] = arr[j], arr[i]
    return arr
```

```python
shufflee(['a', 'b', 'c'])
```
