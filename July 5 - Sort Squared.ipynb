```python
Given a sorted list of integers, square the elements and give the output in sorted order.

For example, given [-9, -2, 0, 2, 3], return [0, 4, 4, 9, 81].
```

```python
def sortsquare(data):
    squared = []
    for i in data:
        square = i**2
        squared.append(square)
    return sorted(squared)
```

```python
sortsquare([-9, -2, 0, 2, 3])
```
