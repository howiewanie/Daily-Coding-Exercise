```python
#Given an integer n and a list of integers l, write a function that randomly
#generates a number from 0 to n-1 that isn't in l (uniform).
```

```python
import random
def check_rand(n,l):
    a = random.randint(1,n-1)
    if a in l:
        a = check_rand(n, l)
    else:
        return a
    print(a)
```

```python
check_rand(9,(3,4,5))
```
