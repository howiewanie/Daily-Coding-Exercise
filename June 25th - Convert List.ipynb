```python
def changenumber(num):
    numlist = [int(x) for x in str(num)]
    newlist = [0,3,6,1,1,1,1,1]
    for n in range(len(numlist)-1):
        if n == 0:
            newlist[0]=numlist[1] 
            newlist[1] = numlist[0]
        elif n%2 == 0:
            newlist[n-1]=numlist[n]
            newlist[n]=numlist[n-1]
        newstring = ' '.join([str(num) for num in newlist]) 
    print(newstring)
```
